# **Z-Test**


Suppose the IQ in a certain population is normally distributed with a mean of  = 100 and standard deviation of  = 15.
A researcher wants to know if a new drug affects IQ levels, so he recruits 20 patients to try it and records their IQ levels.
data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 105, 109, 109, 109, 110, 112, 112, 113, 114, 115]

```python
data2 =  [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 105, 109, 109, 109, 110, 112, 112, 113, 114, 115]
```

**importing z-test**

```python
from statsmodels.stats.weightstats import ztest
```

```python
z_stat,p_value = ztest(data2,value =100)
```

```python
p_value
```

```python
if p_value <.05 :
  print("Reject Null Hypothesis")
else :
   print("Accept Null Hypothesis")
```

# **two sample z test**

cityA = [82, 84, 85, 89, 91, 91, 92, 94, 99, 99, 105, 109, 109, 109, 110, 112, 112, 113, 114, 114]
cityB = [90, 91, 91, 91, 95, 95, 99, 99, 108, 109, 109, 114, 115, 116, 117, 117, 128, 129, 130, 133]

```python
z_stat,p_value = ztest(cityA,cityB)
```

# **T-Test 1sample**

Lets say we want to determine if on average girls score more than 600 in the exam. We do not have the information related to variance (or standard deviation) for girls scores. To perform a t-test, we randomly collect the data of 10 girls with their marks and choose our  value (significance level) to be 0.05 for Hypothesis Testing.
Marks=[587, 602,627,610,619,622,605,608,596,592]

```python
Marks=[587, 602,627,610,619,622,605,608,596,592]
```

```python
from scipy.stats import ttest_1samp
```

```python
t_stat,p_value = ttest_1samp(Marks,600)
```

```python
if p_value <.05 :
  print("Reject Null Hypothesis")
else :
   print("Accept Null Hypothesis")
```

# CHI-SQUARE TEST

A shop owner claims that an equal number of customers come into his shop each weekday. To test this hypothesis, a researcher records the number of customers that come into the shop in a given week and finds the following:
Monday: 50 customers
Tuesday: 60 customers
Wednesday: 40 customers
Thursday: 47 customers
Friday: 53 customers
Perform a Chi-Square goodness of fit test in Python to determine if the data is consistent with the shop owners claim.

```python
expected = [50,50,50,50,50]
observed = [50,60,40,47,53]
```

```python
from scipy.stats import chisquare
```

```python
chi_stat,p_value = chisquare(f_obs = observed,f_exp=expected)
```

```python
p_value
```

```python
if p_value <.05 :
  print("Reject Null Hypothesis")
else :
   print("Accept Null Hypothesis")
```

**Q) The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.**

```python
from scipy.stats import ttest_rel
t_statistic, p_value = ttest_rel(data['Sales_before_digital_add(in $)'], data['Sales_After_digital_add(in $)'])

if p_value < 0.05:
    print("Reject Null Hypothesis : There is a significant increase in sales after stepping into digital marketing.")
else:
    print("Accept Null Hypothesis : There is no significant increase in sales after stepping into digital marketing.")
```

**Q) The company needs to check whether there is any dependency between
thefeatures Region and Manager.**

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(data['Region'], data['Manager'])

print(contingency_table)
print("\n")

# Perform the Chi-Square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the result
if p_value < 0.05:
    print("Reject Null Hypothesis : There is a dependency between the features 'Region' and 'Manager'.")
else:
    print("Accept Null Hypothesis : There is no dependency between the features 'Region' and 'Manager'.")
```

# **ONE WAY AnoVa**

A researcher recruits 30 students to participate in a study. The students are randomly assigned to use one of three studying techniques for the next three weeks to prepare for an exam. At the end of the three weeks, all of the students take the same test.

group1 = [85, 86, 88, 75, 78, 94, 98, 79, 71, 80]

group2 = [91, 92, 93, 85, 87, 84, 82, 88, 95, 96]

group3 = [79, 78, 88, 94, 92, 85, 83, 85, 82, 81]

```python
import numpy as np
import pandas as pd

group1 = [85, 86, 88, 75, 78, 94, 98, 79, 71, 80]

group2 = [91, 92, 93, 85, 87, 84, 82, 88, 95, 96]

group3 = [79, 78, 88, 94, 92, 85, 83, 85, 82, 81]
```

```python
from scipy.stats import f_oneway

f_stat,p_value = f_oneway(group1,group2,group3)
```

```python
p_value
```

```python
if p_value < 0.05:
    print("Reject Null Hypothesis : . ")
else:
    print("Accept Null Hypothesis : studying technique no effect on marks.")
```
