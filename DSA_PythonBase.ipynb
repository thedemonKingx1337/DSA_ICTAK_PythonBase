```python
import pandas as pd

city1=pd.Series([12,-6,8,-3])
print(city1)

print("\n")
print(city1.index)

print("\n")
print(city1.values)

print("\n")
city2=pd.Series([12,-6,8,-3],index=["day1","day2","day3","day4"])
print(city2)

print("\n")
print(city2["day2"])

print("\n")
print(city2[["day2","day3"]])

print("\n")
print(city2>0)

print("\n")
print(city2[city2>0])

print("\n")
print(city2.max())

print("\n")
print(city2.min())

print("\n")
print(city2.mean())

print("\n")
print(city2+5)

print("\n")
place={'Kerala': 28, 'Andhra':25, 'TN':31, 'Karnataka': 19, 'Delhi':38}
place=pd.Series(place)
print(place)


print("\n")
states = ['Andhra', 'Delhi', 'TN', 'MP', 'Kerala', 'Karnataka']
city3 = pd.Series (place, index=states)
print(city3)

print("\n")
print(city3.isna())

print("\n")
print(city3[city3.isna()])

print("\n")
print(city3[city3.notna()])
```

```python
import pandas as pd

data = {'Year': [2001,2002, 2004, 2005, 2006], 'States': ['Kerala', 'TN', 'Andhra', 'Karnataka', 'Bihar'], 'Pop': [300, 250, 280, 400, 360]}
population=pd.DataFrame(data)
print(population)

print("\n")
population_1=pd.DataFrame(data, columns=['Year', 'States', 'Pop', 'Income'])
print(population_1)

print("\n")
print(population_1["Pop"])

print("\n")
print(population_1["Pop"])

print("\n")
population_1['Income']=[20000, 30000,25000, 12000,15000]
print(population_1)

print("\n")
population_1["Debt"]=[320,450,200,150,140]

print("\n")
print(population_1)

print("\n")
population_1['Expense']=population_1['Income']-population_1['Debt']
print(population_1)

print("\n")
population_1=population_1.drop("Expense",axis=1)
print(population_1)

print("\n")
print(population_1.shape)
print(population_1.size)
```
